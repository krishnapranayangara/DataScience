```python
reg_id = ["56","92"," 29"," 93"," 55 ","32"]

reg_dates = ["2021-01-01","2021-02-12", "2021-04-16", "2021-01-22", "2021-01-15", "2021-02-26"]
```

```python
def solve(reg_id, reg_dates):
    res_df = pd.DataFrame(zip(reg_id, reg_dates), columns=["RID", "RDate"])
    """
    input:
    reg_id -> id of the input dataframe
    reg_dates -> dates of the input dataframe
    res_df -> the input dataframe

    output:
    return the resultant dataframe after performing the required operation
    """
    # YOUR CODE GOES HERE
    res_df["RDate"] = res_df["RDate"].astype(str)
    res_df[["RYear","RMonth","RDay"]] = res_df["RDate"].str.split("-",expand=True)
    res_df["RMonth"] = res_df["RMonth"].astype(int)
    res_df["RYear"] = res_df["RYear"].astype(int)
    res_df["RDay"] = res_df["RDay"].astype(int)

    return res_df
```

```python
solve(reg_id, reg_dates)
```
