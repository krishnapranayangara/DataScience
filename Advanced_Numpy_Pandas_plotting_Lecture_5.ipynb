## **1. Introduction to Data Visualization with Matplotlib and Seaborn üöÄ**

**üìö Context:**

In this lecture, we will introduce two of the most popular Python libraries for data visualization: Matplotlib and Seaborn.

You will learn how these tools can help transform complex datasets, like our E-commerce Review Dataset, into insightful visual representations. Let's begin with a solid foundation!

### **1. What is Data Visualization? üìä**

Data visualization is the art of representing data graphically. It allows us to:  
- **Summarize** large datasets efficiently.  
- **Spot patterns, trends, and anomalies**.  
- **Communicate insights to others**.  
- **Support decision-making processes**.  

### Why Do We Need Data Visualization? üßê  
1. **Simplify Complex Data:** Our dataset has 30 columns and 1737 rows. Without visuals, identifying relationships and distributions is challenging.  
2. **Highlight Insights:** For example, **understanding the distribution of ratings** or spotting **outliers in price** is much easier visually.  
3. **Support Hypotheses:** Visuals help validate our assumptions (e.g., Are higher-priced products better rated?).  



### **2. Key Libraries for Visualization**

### Matplotlib üìà  
**Matplotlib** is a versatile library for creating 2D plots and graphs in Python.  
- **Modules:** The `pyplot` module is like the "Swiss Army knife" of visualization‚Äîit provides tools for making basic plots like line graphs, bar charts, and scatter plots.  
- **Customizability:** Highly customizable but can be verbose for more advanced visuals.

### Seaborn üé®  
**Seaborn** is built on top of Matplotlib and is tailored for statistical data visualization.  
- **Enhanced Aesthetics:** Seaborn‚Äôs visuals are clean, professional, and easy to interpret.  
- **Pandas Integration:** It works seamlessly with Pandas DataFrames, making it ideal for datasets like ours.  
- **Advanced Features:** Built-in support for advanced plots like heatmaps, pair plots, and categorical plots.  

üí° **Key Difference:** Matplotlib is great for full control and custom plots, while Seaborn simplifies and enhances plotting for statistical data.

Woah! There is a lot of information in this image. Let's understand them one at a time.

- Figure: The **overall window** or page that everything is drawn on.
  - You can create multiple independent Figures in Jupyter.
  - If you run the code in terminal, separate windows will pop-up.
- Axes: You can add multiple **Axes** to the Figure, which represents a plot.

- **Axis**: Simply the `x-axis` and `y-axis`

- **Axes**: It is the **area** on which the **data is plotted** with functions such as `plot()`.
    - **x-label**: Name of x-axis
    - **y-label**: Name of y-axis

- **Major ticks**:
  - Subdivides the axis into major units.
  - They appear by default during plotting.
- **Minor ticks**:
  - Subdivides the major tick units.
  - They are by default hidden and can be toggled on.

- **Title**: Title of each plot **(Axes)**

- **Subtitle**: The common title of all the plots.

- **Legend**:
  - Describes the elements in the plot.
  - Blue and Green curves in this case.

These are the major components of a matplotlib plot.

### **4. How to Choose the Right Plot? ü§î**  

Choosing the right plot depends on your **question of interest** and the nature of your data. Before selecting a visualization:  

1. **What is the question you want to answer?**  
2. **How many variables are involved?**  
3. **Are the variables numerical or categorical?**


### Step 1: Number of Variables Involved  
The number of variables determines the type of analysis:  

- **1 Variable** ‚Üí **Univariate Analysis**  
- **2 Variables** ‚Üí **Bivariate Analysis**  
- **3+ Variables** ‚Üí **Multivariate Analysis**


## **2. Univariate Data Visualization üé®**  

We start with univariate plots to understand **distributions** and **frequencies** of single columns.






### 1 üìä Visualization Goal: Histogram
A **histogram** helps visualize the distribution of a numerical column. Let‚Äôs explore the distribution of <font color="green">`rating`</font>.

#### Key Insights
- Identify skewness (are ratings generally high or low?).
- Check for outliers or unusual spikes in values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for 'rating'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rating', kde=False, color='skyblue')

# Styling Tip üí°
# Add a vertical line for the mean rating
mean_rating = df['rating'].mean()
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")

plt.title("Distribution of Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()
```

Note: Equivalent in <font color="skyblue">Matplotlib</font>:

```python
plt.figure(figsize=(8, 6))
plt.hist(df['rating'].dropna(), bins=10, color='skyblue')
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

### 2 üìä Visualization Goal: Bar Plot (Categorical)

A bar plot compares different categories, such as the count of products in each <font color="green">category</font>.

**Key Insights**
- Identify popular product categories.
 -Useful for visualizing categorical data frequencies.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='category', palette='coolwarm')

# Styling Tip üí°
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.title("Count of Products by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

Note: Equivalent in <font color="greed">Matplotlib</font>:

```python
categories = df['category'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(categories.index, categories.values, color='skyblue')
plt.xticks(rotation=45)
plt.title("Count of Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

### **3. üìä Visualization Goal: KDE (Kernel Density Estimate) Plot**

A KDE plot is a smoothed version of the histogram. Let‚Äôs examine the distribution of <font color="green">discounted_price</font>.

**Key Insights**
- Easily see peaks and valleys in the distribution.
- Compare with a histogram for a smoothed perspective.

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='discounted_price', shade=True, color='purple')

# Styling Tip üí°
# Add a vertical grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.title("KDE Plot of Discounted Prices", fontsize=16)
plt.xlabel("Discounted Price", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.show()
```

Note: In <font color="greed">Matplotlib</font>, you could combine plt.hist with density=True and then plot a plt.plot of a KDE function from scipy.stats if needed.

## **3. Bivariate Plots (Two Variables)**

These plots help us understand relationships between two variables.

### **1 üìä Visualization Goal: Scatter Plot**

A scatter plot shows the relationship between two numerical variables. Let‚Äôs see if there‚Äôs a correlation between <font color="green">discounted_price</font> and <font color="green">rating</font>.

**Key Insights**
- Detect patterns (does a higher price correlate with higher ratings?).
- Identify outliers (very high price but low rating, for example).

```python
plt.figure(figsize=(8, 6))

# Styling Tip üí°
# Use alpha to handle overlapping points
sns.scatterplot(data=df, x='discounted_price', y='rating', color='teal', alpha = 0.8)

plt.title("Discounted Price vs. Rating", fontsize=16)
plt.xlabel("Discounted Price", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.show()
```

### **2. üìä Visualization Goal: Line Plot**

A line plot often represents data over time. Let‚Äôs plot average <font color="green">rating_filled</font> by <font color="green">month_year</font> to see trends.

**Key Insights**
- Observe how ratings change month to month.
- Could highlight seasonal spikes or dips.

```python
# Group by 'month_year' and calculate average rating
monthly_rating = df.groupby('month_year')['rating_filled'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(monthly_rating['month_year'], monthly_rating['rating_filled'], marker='o', color='navy')

# Styling Tip üí°
# Rotate the x-axis labels and add a grid
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)

plt.title("Average Rating by Month-Year", fontsize=16)
plt.xlabel("Month-Year", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.tight_layout()
plt.show()
```

Note: You can use <font color="magenta">Seaborn</font>‚Äôs lineplot:

```python
sns.lineplot(data=monthly_rating, x='month_year', y='rating_filled', marker='o', color='navy')
plt.xticks(rotation=45)
plt.title("Average Rating by Month-Year")
plt.xlabel("Month-Year")
plt.ylabel("Average Rating")
plt.show()
```

### **3. üìä Visualization Goal: Box Plot**

A box plot shows the distribution of a numerical variable across different categories. Let‚Äôs examine <font color="green">rating</font> by <font color="green">category</font>.

**Key Insights**
- Spot differences in median rating by category.
- Identify potential outliers within each category.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='category', y='rating', palette='Set3') # Styling Tip üí° palette

plt.xticks(rotation=45)
plt.title("Rating Distribution by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.show()
```

Note: <font color="greed">Matplotlib</font> has plt.boxplot, for similar plot

### **4. üìä Visualization Goal: Violin Plot**

A violin plot is similar to a box plot but also shows the kernel density of the data. Let‚Äôs see the distribution of <font color="green">discounted_price</font> by <font color="green">rating_category</font>.

**Key Insights**
- Compare the spread of prices for different rating categories.
- Check if there's a concentration of lower/higher prices in certain categories.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='rating_category', y='discounted_price', inner='quartile', palette='Accent')

plt.title("Violin Plot of Discounted Price by Rating Category", fontsize=16)
plt.xlabel("Rating Category", fontsize=12)
plt.ylabel("Discounted Price", fontsize=12)
plt.show()
```

### **5. üìä Visualization Goal: Grouped Bar Plot**

A grouped bar plot compares multiple categories side by side. Let‚Äôs compare the average <font color="green">rating</font> for products that do or do not contain the keyword "durable", across different <font color="green">rating_category</font>.

**Key Insights**
- Check how ‚Äúdurable‚Äù items perform in each rating category vs. non-durable.
- Identify which group has higher/lower average ratings.

```python
# Create a pivoted dataframe
grouped_data = df.groupby(['rating_category', 'contains_durable'])['rating_filled'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='rating_category', y='rating_filled', hue='contains_durable', palette='Paired')

# Styling Tip üí°
# Add legend title
plt.legend(title='Contains Durable?')

plt.title("Average Rating by Rating Category and Durability", fontsize=16)
plt.xlabel("Rating Category", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.show()
```

## **4. Multivariate Plots (Three or More Variables)**


### **1. üìä Visualization Goal: Pair Plot**

A pair plot visualizes pairwise relationships and distributions for multiple numerical variables. Let‚Äôs look at <font color="green">rating</font>, <font color="green">discounted_price</font>, and <font color="green">actual_price</font>.

**Key Insights**
- Quickly spot correlations.
- View distributions on the diagonal.

```python
sns.pairplot(df[['rating', 'discounted_price', 'actual_price', 'calculated_discount_filled']].dropna(),
             diag_kind='kde', corner=True, plot_kws={'alpha':0.5})
plt.show()
```

### **2. üìä Visualization Goal: Heatmap**

A heatmap is great for correlation matrices. Let‚Äôs see how <font color="green">rating</font>, <font color="green">discounted_price</font>, <font color="green">actual_price</font>, <font color="green">rating_count</font>, etc., correlate.

**Key Insights**
- Identify strong positive or negative correlations.
- Guides feature selection or further analysis.

```python
corr_cols = ['rating', 'discounted_price', 'actual_price', 'rating_count', 'calculated_discount']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(8, 6))

# Styling Tip üí°
# Format annotation to 2 decimal places
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='Blues')

plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

### **3. Subplots and hue**

The **`hue`** parameter in <font color="magenta">Seaborn</font> allows us to color the data points (or bars, lines, etc.) by another categorical variable, giving a quick comparison without creating separate subplots or separate figures.

**Example: Bar Plot with Hue**

Let‚Äôs say we want to compare the **average rating** by **`rating_category`** for products that **do or do not contain ‚Äúdurable‚Äù**. We'll use the `contains_durable` column as the `hue` parameter.


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Group by rating_category and contains_durable, then compute average rating
grouped_data = df.groupby(['rating_category', 'contains_durable'])['rating_filled'].mean().reset_index()

sns.barplot(
    data=grouped_data,
    x='rating_category',
    y='rating_filled',
    hue='contains_durable',
    palette='viridis'
)

# Styling Tip üí°
# Add a legend title and rotate x-axis labels
plt.legend(title='Durable?')
plt.xticks(rotation=45)
plt.title("Average Rating by Rating Category and Durability", fontsize=16)
plt.xlabel("Rating Category", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.show()
```

**Subplots with Different Hues**

Sometimes, you may want multiple subplots side by side, each using a different hue variable. This can be useful if you have multiple categorical variables you want to highlight in separate views.

**Example:** Two Scatter Plots with Different Hues
- In the first subplot, we'll color by <font color="green">rating_category</font>.
- In the second subplot, we‚Äôll color by the <font color="green">price_range</font> column (imagine it categorizes items as ‚ÄúBudget‚Äù, ‚ÄúMid-range‚Äù, ‚ÄúPremium‚Äù, etc.).

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Subplot 1: Hue by 'rating_category'
sns.scatterplot(
    data=df,
    x='discounted_price',
    y='rating_filled',
    hue='rating_category',
    ax=axes[0],
    palette='coolwarm'
)
axes[0].set_title("Discounted Price vs. Rating (Hue: Rating Category)", fontsize=14)
axes[0].set_xlabel("Discounted Price", fontsize=12)
axes[0].set_ylabel("Rating", fontsize=12)

# Subplot 2: Hue by 'price_range'
sns.scatterplot(
    data=df,
    x='discounted_price',
    y='rating_filled',
    hue='price_range',  # Alternatively: 'actual_price', binned, or another categorical col
    ax=axes[1],
    palette='Spectral'
)
axes[1].set_title("Discounted Price vs. Rating (Hue: Price Range)", fontsize=14)
axes[1].set_xlabel("Discounted Price", fontsize=12)
axes[1].set_ylabel("Rating", fontsize=12)

plt.tight_layout()
plt.show()
```

### **4. Stacked Plots**

A stacked plot (most commonly a stacked bar plot) allows you to see parts of a whole for different categories on top of each other.

**1 Stacked Bar Plot**
Let‚Äôs say we want to visualize how rating_category is distributed within each <font color="green">price_range</font> category. This will show each price_range on the x-axis and ‚Äústack‚Äù the segments of rating categories in each bar.

Steps:
- Compute counts of reviews (or any metric) grouped by price_range and rating_category.
- Unstack the grouped data so each rating_category becomes a separate column.
- Use <font color="skyblue">Matplotlib</font> (pandas plotting) with stacked=True.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Group by price_range and rating_category, and get counts
grouped_stack = df.groupby(['price_range', 'rating_category'])['review_id'].count().unstack()

# 2. Plot as stacked bar
grouped_stack.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='Blues'
)

# Styling Tip üí°
plt.title("Stacked Bar Plot: Rating Categories by Price Range", fontsize=16)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Number of Reviews", fontsize=12)
plt.legend(title="Rating Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Note: A stacked bar plot can become hard to read if there are many categories. In such cases, a grouped bar chart or a different type of visualization might be more suitable.
